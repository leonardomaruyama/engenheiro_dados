```python
import os
import glob

# 🔥 Caminho para a pasta de dados
diretorio = r'C:\Users\Leonardo\OneDrive\Documentos\Git\teste_engenheiro\eng_dados\raw'

# 🔥 Busca e remove os arquivos usando um padrão
arquivos = glob.glob(os.path.join(diretorio, '*.parquet'))

for arquivo in arquivos:
    os.remove(arquivo)
    print(f"✅ Arquivo removido: {arquivo}")

print(f"🚀 Removidos {len(arquivos)} arquivos com o padrão *.parquet")
```
